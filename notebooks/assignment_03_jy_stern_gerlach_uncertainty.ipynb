## Recap: Key Results from the Previous Notebook

The following results were established in the previous notebook. They are collected here for reference.

**Spin-1 operators in the $\tilde{J}_z$ basis** (dimensionless, scaled by $\hbar$, so $J_i = \hbar\tilde{J}_i$):

$$
\tilde{J}_x = \frac{1}{\sqrt{2}}
\begin{pmatrix}0&1&0\\1&0&1\\0&1&0\end{pmatrix}, \qquad
\tilde{J}_y = \frac{1}{\sqrt{2}}
\begin{pmatrix}0&-i&0\\i&0&-i\\0&i&0\end{pmatrix}, \qquad
\tilde{J}_z = \begin{pmatrix}1&0&0\\0&0&0\\0&0&-1\end{pmatrix}
$$

**The Casimir operator** $\tilde{J}^2 = \tilde{J}_x^2 + \tilde{J}_y^2 + \tilde{J}_z^2 = 2I$, confirming $j = 1$.

**The $\tilde{J}_z$ eigenstates** (standard basis):

$$
|1,1\rangle \to \begin{pmatrix}1\\0\\0\end{pmatrix}, \qquad
|1,0\rangle \to \begin{pmatrix}0\\1\\0\end{pmatrix}, \qquad
|1,-1\rangle \to \begin{pmatrix}0\\0\\1\end{pmatrix}
$$

satisfying $\tilde{J}_z|1,m\rangle = m|1,m\rangle$.

**Ladder operators:**

$$
\tilde{J}_+ = \begin{pmatrix}0&\sqrt{2}&0\\0&0&\sqrt{2}\\0&0&0\end{pmatrix}, \qquad
\tilde{J}_- = \begin{pmatrix}0&0&0\\\sqrt{2}&0&0\\0&\sqrt{2}&0\end{pmatrix}
$$

**Commutation relations** (verified numerically):

$$
[\tilde{J}_x, \tilde{J}_y] = i\tilde{J}_z, \qquad
[\tilde{J}_y, \tilde{J}_z] = i\tilde{J}_x, \qquad
[\tilde{J}_z, \tilde{J}_x] = i\tilde{J}_y
$$

```python
import sympy as sp
import matplotlib.pyplot as plt

sp.init_printing()

I = sp.I
sqrt2 = sp.sqrt(2)

# Spin-1 operators (dimensionless, J_i = hbar * J_t_i)
Jx_t = (1/sqrt2) * sp.Matrix([
    [0,  1,  0],
    [1,  0,  1],
    [0,  1,  0]
])

Jy_t = (1/sqrt2) * sp.Matrix([
    [ 0, -I,  0],
    [ I,  0, -I],
    [ 0,  I,  0]
])

Jz_t = sp.Matrix([
    [1, 0,  0],
    [0, 0,  0],
    [0, 0, -1]
])

J2_t = Jx_t**2 + Jy_t**2 + Jz_t**2

Jp_t = Jx_t + I * Jy_t
Jm_t = Jx_t - I * Jy_t

# J_z basis states
ket_p1 = sp.Matrix([1, 0, 0])
ket_0  = sp.Matrix([0, 1, 0])
ket_m1 = sp.Matrix([0, 0, 1])

# Commutator
def comm(A, B):
    return A*B - B*A

# Expectation value <ket|op|ket>
def expect(op, ket):
    return sp.simplify((ket.H * op * ket)[0, 0])

# Uncertainty sqrt(<op^2> - <op>^2)
def uncertainty(op, ket):
    exp1 = expect(op, ket)
    exp2 = expect(op**2, ket)
    return sp.simplify(sp.sqrt(exp2 - exp1**2))

print("Setup complete.")
```

## Part 3: Uncertainty Relations for Angular Momentum

In Section 3.5, Townsend derives the general uncertainty relation for two 
non-commuting Hermitian operators $\hat{A}$ and $\hat{B}$ satisfying $[\hat{A}, \hat{B}] = i\hat{C}$:

$$
\Delta A \cdot \Delta B \geq \frac{|\langle C \rangle|}{2}
$$

where $\Delta A = \sqrt{\langle A^2 \rangle - \langle A \rangle^2}$.

Since $[\hat{J}_x, \hat{J}_y] = i\hbar\hat{J}_z$, the angular momentum uncertainty relation is

$$
\Delta J_x \cdot \Delta J_y \geq \frac{\hbar}{2}|\langle J_z \rangle|.
$$

In terms of our dimensionless operators $\tilde{J}_i = J_i/\hbar$ this becomes

$$
\Delta\tilde{J}_x \cdot \Delta\tilde{J}_y \geq \frac{1}{2}|\langle \tilde{J}_z \rangle|.
$$

```python
# Verify your hand calculation for |1,1>
print("State |1,1>")
print("─" * 40)

dJx    = uncertainty(Jx_t, ket_p1)
dJy    = uncertainty(Jy_t, ket_p1)
Jz_exp = expect(Jz_t, ket_p1)
lhs    = sp.simplify(dJx * dJy)
rhs    = sp.simplify(sp.Abs(Jz_exp) / 2)

print(f"ΔJ̃x          = {dJx}")
print(f"ΔJ̃y          = {dJy}")
print(f"ΔJ̃x · ΔJ̃y   = {lhs}")
print(f"½|<J̃z>|      = {rhs}")
print(f"LHS - RHS    = {sp.simplify(lhs - rhs)}")
```
